# **India**

```python
df_india = df[df['location'] == 'India']
df_india.head()
```

```python
df_india.fillna(0,inplace=True)
```

```python
df_india = df_india.rename(columns={'date':'ds','new_cases':'y'})
```

```python
model1 = Prophet(changepoint_prior_scale=0.2)
model1.fit(df_india)
x1 = model1.make_future_dataframe(periods= 30, freq='D')
forecast1 = model1.predict(x1)
df3 = forecast1[['ds','trend','yhat_lower','yhat_upper','yhat']]
model1.plot(df3,figsize=(20,5));
```

# **Armenia**

```python
df_Armenia = df[df['location'] == 'Armenia']
df_Armenia.fillna(0,inplace=True)
df_Armenia = df_Armenia.rename(columns={'date':'ds','new_cases':'y'})
model2 = Prophet(changepoint_prior_scale=0.2)
model2.fit(df_Armenia)
x2 = model2.make_future_dataframe(periods= 30, freq='D')
forecast2 = model2.predict(x2)
df4 = forecast2[['ds','trend','yhat_lower','yhat_upper','yhat']]
model2.plot(df4,figsize=(20,5));
```
